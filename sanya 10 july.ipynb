```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 6))

x_body = [2, 2, 6, 6, 2]
y_body = [2, 5, 5, 2, 2]
plt.plot(x_body, y_body, color='red')

x_roof = [2, 4, 6]
y_roof = [5, 7, 5]
plt.plot(x_roof, y_roof, color='red')

x_door = [3.5, 3.5, 4.5, 4.5, 3.5]
y_door = [2, 4, 4, 2, 2]
plt.plot(x_door, y_door, color='brown')

x_window1 = [2.5, 2.5, 3.2, 3.2, 2.5]
y_window1 = [3.5, 4.2, 4.2, 3.5, 3.5]
plt.plot(x_window1, y_window1, color='pink')

x_window2 = [4.8, 4.8, 5.5, 5.5, 4.8]
y_window2 = [3.5, 4.2, 4.2, 3.5, 3.5]
plt.plot(x_window2, y_window2, color='pink')

plt.plot([2.85, 2.85], [3.5, 4.2], color='yellow')
plt.plot([2.5, 3.2], [3.85, 3.85], color='yellow')

plt.plot([5.15, 5.15], [3.5, 4.2], color='yellow')
plt.plot([4.8, 5.5], [3.85, 3.85], color='yellow')

plt.plot(4.3, 3, 'ko')#door handle

plt.xlim(0, 8)
plt.ylim(0, 8)
plt.gca().set_aspect('equal')
plt.title("2D House")
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arc

fig, ax = plt.subplots(figsize=(6,6))

# Face
face = Circle((0,0), 1, color='yellow', ec='black')
ax.add_patch(face)

# Heart Eyes
ax.text(-0.45, 0.25, "❤", fontsize=28, color="red")
ax.text(0.20, 0.25, "❤", fontsize=28, color="red")

# Smile
smile = Arc((0,-0.1), 1.2, 0.8, theta1=200, theta2=340,
            linewidth=3, color='black')
ax.add_patch(smile)

ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
ax.set_aspect('equal')


plt.show()
```
