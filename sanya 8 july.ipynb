```python
import matplotlib.pyplot as plt
#line chart
x=[1,2,3,4]
y=[10,15,12,20]
plt.plot(x,y,marker='o')
plt.show()
```

```python
#bar chart
students=["a","b","c","d"]
marks=[80,90,75,60]
plt.bar(students,marks)
plt.show()
```

```python
#horizontal bar
plt.barh(students,marks)
plt.show()
```

```python
#histogram
data=[10,12,15,18,20,21,22,24,25]
plt.hist(data,bins=5)
plt.show()
```

```python
#scatter plot
x=[1,2,3,4]
y=[5,8,6,10]
plt.scatter(x,y)
plt.show()
```

```python
#pie chart
labels=["a","b","c"]
values=[40,35,25]
plt.pie(values,labels=labels,autopct="%1.1f%%")
plt.show()
```

```python
#boxplot
data=[10,20,30,40,100]
plt.boxplot(data)
plt.show()
```

```python
#violin plot
data=[[10,20,30,40,50]]
plt.violinplot(data)
plt.show()
```

```python
#area chart
x=[1,2,3,4]
y=[2,4,3,5]
plt.fill_between(x,y)
plt.show()
```

```python
#stack plot
x=[1,2,3]
y1=[2,3,4]
y2=[1,2,3]
plt.stackplot(x,y1,y2)
plt.show()
```

```python
#stem plot
x=[1,2,3]
y=[2,5,3]
plt.stem(x,y)
plt.show()
```

```python
#error bar
x=[1,2,3]
y=[5,6,7]
e=[0.5,0.3,0.4]
plt.errorbar(x,y,yerr=e)
plt.show()
```

```python
#hexbin
plt.hexbin(x,y,gridsize=10)
plt.show()
```

```python
#evemt plot
data=[[1,2,3],[2,4,6]]
plt.eventplot(data)
plt.show()
```
