```python
#polar plot
import numpy as np
import matplotlib.pyplot as plt
t=np.linspace(0,6.28,100)
r=t
plt.polar(t,r)
plt.show()
```

```python
#contour plot
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
X,Y=np.meshgrid(x,y)
Z=X**2+Y**2
plt.contour(X,Y,Z)
plt.show()
```

```python
#filled contour
plt.contourf(X,Y,Z)
plt.show()
```

```python
#image plot
import matplotlib.image as mpimg
img=mpimg.imread("shortcake.jpg")
plt.imshow(img)
plt.show()
```

```python
#heatmap
data=[[1,2],[3,4]]
plt.imshow(data)
plt.colorbar()
plt.show()
```

```python
#3d surface
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=fig.add_subplot(111,projection="3d")
ax.plot_surface(X,Y,Z)
```
