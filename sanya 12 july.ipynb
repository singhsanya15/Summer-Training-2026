```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arc

fig, ax = plt.subplots(figsize=(6,6))

# Face
face = Circle((0,0), 1, fill=True, color="yellow", linewidth=3)
ax.add_patch(face)

# Eyes
ax.plot([-0.5,-0.25],[0.45,0.25], linewidth=3,color="black")
ax.plot([0.25,0.5],[0.25,0.45], linewidth=3,color="black")

# Laughing Mouth
mouth = Arc((0,-0.1),1.2,0.9,theta1=200,theta2=340,linewidth=3)
ax.add_patch(mouth)

# Tears
ax.plot([-0.6,-0.55],[0.2,-0.1], linewidth=3,color="blue")
ax.plot([0.6,0.55],[0.2,-0.1], linewidth=3,color="blue")

ax.set_xlim(-1.5,1.5)
ax.set_ylim(-1.5,1.5)
ax.set_aspect('equal')

ax.grid(True)

plt.title("Laughing Emoji")
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arc

# Create figure
fig, ax = plt.subplots(figsize=(6,6))

# Face
face = Circle((0, 0), 1, fill=True,color="yellow", linewidth=3)
ax.add_patch(face)

# Heart Eyes
ax.text(-0.35, 0.35, "❤", fontsize=30, color='red',
        ha='center', va='center')
ax.text(0.35, 0.35, "❤", fontsize=30, color='red',
        ha='center', va='center')

# Eyelashes
# Left
ax.plot([-0.45,-0.52],[0.48,0.60], color='black', linewidth=2)
ax.plot([-0.35,-0.35],[0.50,0.63], color='black', linewidth=2)
ax.plot([-0.25,-0.18],[0.48,0.60], color='black', linewidth=2)

# Right
ax.plot([0.25,0.18],[0.48,0.60], color='black', linewidth=2)
ax.plot([0.35,0.35],[0.50,0.63], color='black', linewidth=2)
ax.plot([0.45,0.52],[0.48,0.60], color='black', linewidth=2)

# Rosy Cheeks
left_cheek = Circle((-0.5, -0.15), 0.12, color='pink', alpha=0.7)
right_cheek = Circle((0.5, -0.15), 0.12, color='pink', alpha=0.7)
ax.add_patch(left_cheek)
ax.add_patch(right_cheek)

# Smile
smile = Arc((0, -0.05), 1.0, 0.8,
            theta1=200, theta2=340,
            linewidth=3)
ax.add_patch(smile)

# Small Hearts Around Face
ax.text(-1.15, 0.95, "❤", fontsize=25, color='pink')
ax.text(1.0, 1.0, "❤", fontsize=25, color='pink')
ax.text(0.95, -1.3, "❤", fontsize=25, color='pink')

# Settings
ax.set_xlim(-1.6, 1.6)
ax.set_ylim(-1.6, 1.6)
ax.set_aspect('equal')

ax.grid(True)
plt.title("Heart Blushing Emoji")

plt.show()
```
