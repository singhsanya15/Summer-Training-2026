```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arc

# Create figure
fig, ax = plt.subplots(figsize=(6,6))

# Face
face = Circle((0, 0), 1, fill=True,color="yellow", linewidth=3)
ax.add_patch(face)

# Left Eye
left_eye = Circle((-0.35, 0.35), 0.08, color='black')
ax.add_patch(left_eye)

# Right Eye
right_eye = Circle((0.35, 0.35), 0.08, color='black')
ax.add_patch(right_eye)

# Smile
smile = Arc((0, -0.1), 1.1, 0.8, theta1=200, theta2=340, linewidth=3)
ax.add_patch(smile)
left_cheek = Circle((-0.75, -0.1), 0.12, color='red', alpha=0.7)
ax.add_patch(left_cheek)
right_cheek = Circle((0.75, -0.1), 0.12, color='red', alpha=0.7)
ax.add_patch(right_cheek)

# Set limits
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

# Equal scaling
ax.set_aspect('equal')
ax.grid(True)

# Title
plt.title("Smiley Emoji")
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arc

fig, ax = plt.subplots(figsize=(6,6))

# Face
face = Circle((0,0), 1, fill=True, color="yellow",linewidth=3)
ax.add_patch(face)

# Heart Eyes
ax.text(-0.35, 0.35, "❤", fontsize=35, ha='center', va='center', color='red')
ax.text(0.35, 0.35, "❤", fontsize=35, ha='center', va='center', color='red')

# Smile
smile = Arc((0,-0.15), 1.1, 0.8, theta1=200, theta2=340, linewidth=3)
ax.add_patch(smile)
left_cheek = Circle((-0.75, -0.1), 0.12, color='pink', alpha=0.7)
ax.add_patch(left_cheek)
right_cheek = Circle((0.75, -0.1), 0.12, color='pink', alpha=0.7)
ax.add_patch(right_cheek)

ax.set_xlim(-1.5,1.5)
ax.set_ylim(-1.5,1.5)
ax.set_aspect('equal')

ax.grid(True)

plt.title("Heart Eyes Emoji")
plt.show()
```
